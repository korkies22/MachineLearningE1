```python
#Importar librerías
import math
import csv
import numpy as np
import pandas as pd
from sklearn.neural_network  import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from random import randrange
from IPython.display import HTML, display
```

```python
#Leer el archivo con los datos y cargarlos en las variables
y = []
x=[]
with open('./data/xtrain.txt') as f:
    for line in f:
        x.append([float(i) for i in line.split()])
with open('./data/ytrain.txt') as g:
    for line in g:
        y.append([float(i) for i in line.split()])
```

Si se utilizan 1000 datos para realizar las validaciones y evaluaciones del modelo respectivamente, entonces la confianza con la que el error estimado esté máximo a 0.1 de diferencia del error real es de:

```python
#La confianza se obtiene de acuerdo a la cuotas de
confianza=1-2*math.exp(-2*0.05**2*1100)
print(confianza)
```

Como se puede observar esta confianza es ligeramente superior al 99%. Lo cual significa que usar 1100 datos es un buen estimativo para estar máximo a 0.05 del error.
Se parten los datos en entrenamiento y test

```python
#Se divide el archivo para entrenamiento y test. 
#Se reserven 600 datos para test de acuerdo con la confianza anteriormente mencionada
xTrain, xValidation, yTrain, yValidation = train_test_split(x, y, test_size = 1100, random_state = 0)
```

```python
#Funciones de utilidad para blanquear los datos
def center(X):
    newX = X - np.mean(X, axis = 0)
    return newX
def whiten(X):
    newX = center(X)
    cov = X.T.dot(X)/float(X.shape[0])
    # Calculate the eigenvalues and eigenvectors of the covariance matrix
    eigVals, eigVecs = np.linalg.eig(cov)
    # Apply the eigenvectors to X
    decorrelated = X.dot(eigVecs)
    # Rescale the decorrelated data
    whitened = decorrelated / np.sqrt(eigVals + 1e-5)
    return whitened
```

```python
#Clase que normaliza las entradas
scaler = preprocessing.StandardScaler()
# Fit your data on the scaler object
x1 = scaler.fit_transform(xTrain)

x2=[]
x2=whiten(np.array(xTrain))
#Se crea un arreglo con los datos crudos, datos normalizados y datos blanqueados
xs=[xTrain,x1,x2]
```

```python
#Función para entrenar una red de una sola capa escondida, 
#La función recibe también como parámetro la función de activación que se desea en las neuronas de capa escondida
def trainNetwork(n,xTrainT,yTrainT,activationF):
    #Se escoge relu puesto que esta fue la que mejor dió en las pruebas realizadas
    clf = MLPClassifier(solver='lbfgs',batch_size=1, alpha=0,hidden_layer_sizes=(n),activation=activationF, max_iter=500, random_state=1)
    clf.fit(xTrainT, yTrainT)   
    return [clf,clf.score(xValidation, yValidation)]
```

```python
#Función que entrena neuronas por tipo de datos y número de neuronas
def trainActiv(activation):
    errors=[]
    neurons=[]
    #Se itera por tipo de dato de entrada de acuerdo al preprocesamiento
    for i in range(len(xs)):
        #Se crean arreglos temporales, cada uno corresponde a un tipo de dato de preprocesamiento
        errorsTemp=[None]
        neuronsTemp=[]
        #Se itera de 1 neurona hasta 50
        for j in range(1, 50):
            #Se almacenan los resultados en los arreglos
            neuron=trainNetwork(j,xs[i],yTrain,activation)
            neuronsTemp.append(neuron[0])
            errorsTemp.append(neuron[1])
        #Al final se concatenan estos arreglos temporales a los globales
        errors.append(errorsTemp)
        neurons.append(neuronsTemp)
    #Se convierten a np
    errors=np.array(errors)
    neurons=np.array(neurons)
    return [errors,neurons]
```

```python
#Se entrenan diferentes tipos de neurona dependiendo de cada función de activación
train1=trainActiv('relu')
train2=trainActiv('logistic')
train3=trainActiv('tanh')
train4=trainActiv('identity')
```

```python
#Algoritmo para encontrar los índices del elemento más grande de un arreglo 2D, en caso de varias ocurrencias retorna la última
def findMaxIndexes(arr):
    maxC=0
    maxIndex=[-1,-1]
    for i in range(len(arr)):
        for j in range(len(arr[0])):
            if arr[i][j] is not None and arr[i][j]>=maxC:
                maxIndex=[i,j]
                maxC=arr[i][j]
    return maxIndex
```

```python
#Se utiliza el algoritmo anterior para encontrar la red de mejor resultado
[ind1,ind2]=findMaxIndexes(train1[0])
print(ind1,ind2)
print(train1[0][ind1][ind2])

[ind1,ind2]=findMaxIndexes(train2[0])
print(ind1,ind2)
print(train2[0][ind1][ind2])

[ind1,ind2]=findMaxIndexes(train3[0])
print(ind1,ind2)
print(train3[0][ind1][ind2])

[ind1,ind2]=findMaxIndexes(train4[0])
print(ind1,ind2)
print(train4[0][ind1][ind2])
```

Arriba se puede observar que la red de mayor porcentaje de éxito es la que maneja Relu como función de activación. Además se observa que de todas las funciones de activación, el menor porcentaje de error se encuentra con datos de entrenamiento que no han sido blanqueados ni normalizados. Esto puede deberse a que en este caso, estos datos crudos ofrecen una mayor expresividad que el modelo aprovecha. Sin embargo, de todas maneras el porcentaje de acierto es bajo, no llega al 70% en ningún caso. Posteriormente se espera que este puntaje aumente. A partir de ahora siempre se utilizará relu y datos crudos para entrenar

```python
#Función para entrenar redes neuronales con distintas tasas de aprendizaje
#Esto se realizará para saber cual es la mejor tasa de aprendizaje a usar.
def trainNetworkLR(learningRate,n):
    lr=0.001*10**learningRate
    clf = MLPClassifier(solver='lbfgs',batch_size=1, alpha=0,hidden_layer_sizes=(n),activation='relu', max_iter=500, learning_rate_init=lr ,random_state=1)
    clf.fit(xTrain, yTrain)   
    return [clf,clf.score(xValidation, yValidation)]
```

```python
#Se itera por tasa de aprendizaje desde 0.0001 hasta 0.1 y desde 1 hasta 30 neuronas
errors=[]
neurons=[]
for i in range(-1,3):
    #Se crean arreglos temporales, cada uno corresponde a una tasa de aprendizaje
    errorsTemp=[None]
    neuronsTemp=[]
    for j in range(1, 30):
        #Se almacenan los resultados en los arreglos
        neuron=trainNetworkLR(i,j)
        neuronsTemp.append(neuron[0])
        errorsTemp.append(neuron[1])
    #Al final se concatenan estos arreglos temporales a los globales
    errors.append(errorsTemp)
    neurons.append(neuronsTemp)
#Se convierten a np
errors=np.array(errors)
neurons=np.array(neurons)
```

```python
#Se busca la neurona con mejor resultado
[ind1,ind2]=findMaxIndexes(errors)
print(ind1,ind2)
print(errors[ind1][ind2])
```

Se puede observar en los resultados anteriores que con una tasa de aprendizaje de 0.1 se obtuvo el modelo de mejor valor, esto es importante ya que significa que se puede utilizar esta tasa y no la que se tiene por defecto para entrenar las redes y por tanto el modelo convergería más rápido. A partir de ahora se utilizará una tasa de aprendizaje de 0.1

```python
#Función para entrenar redes neuronales de distintas capas y distinto número de neuronas
#Para facilidad todas las capas tendrán el mismo número de neuronas
#Esta función además calcula la tasa de éxito de clasificación con los datos de validación.
#Se retorna la red y esta tasa de éxito
#Los otros parámetros de la red son similares a los del caso de una sola capa
def trainNetworkLayers(nLayers,nNeurons):
    a = (nNeurons,)*nLayers
    clf = MLPClassifier(solver='lbfgs',batch_size =1, alpha=0,hidden_layer_sizes=a,activation="relu", max_iter=500, learning_rate_init=0.1, random_state=1)
    clf.fit(xTrain, yTrain)   
    return [clf,clf.score(xValidation,yValidation)]
```

```python
#Se itera por número de capas y número de neuronas, desde una capa hasta 10, y desde una neurona por capa hasta 30
errorsLayers=[]
neurons2=[]
for i in range(1, 20):
    #Se crean arreglos temporales, cada uno corresponde a un número de capas
    errorsTemp=[None]
    neuronsTemp=[]
    for j in range(1, 50):
        #Se almacenan los resultados en los arreglos
        neuron=trainNetworkLayers(i,j)
        neuronsTemp.append(neuron[0])
        errorsTemp.append(neuron[1])
    #Al final se concatenan estos arreglos temporales a los globales
    errorsLayers.append(errorsTemp)
    neurons2.append(neuronsTemp)
#Se convierten a np
errorsLayers=np.array(errorsLayers)
neurons2=np.array(neurons2)
```

```python
#Se utiliza el algoritmo anterior para encontrar la red de mejor resultado
[ind1,ind2]=findMaxIndexes(errorsLayers)
print(ind1,ind2)
print(errorsLayers[ind1][ind2])
```

Los resultados de arriba indican que una red con 7 capas de 10 neuronas cada una tiene el mejor resultado. De acuerdo con los cálculos anteriores, se puede afirmar con una confianza de 99% que el porcentaje de clasificación exitoso es mínimo de 67.5% (0.68-0.05 al ser tasa de clasificación exitosa === 1-error). Este resultado no corresponde a un valor muy alto (no garantiza que clasifique 7 o más correctamente de cada 10 datos). Sin embargo fue el mejor resultado obtenido y por tanto este es el que se usará. 

```python
#Cargar los datos finales para realizar la predicción correspondiente
xFinal=[]
with open('./data/xtest.txt') as f:
    for line in f:
        xFinal.append([float(i) for i in line.split()])
```

```python
bestNN= neurons2[ind1][ind2]
#Esta red se entrena con los datos de entrenamiento totales
bestNN.fit(x, y)
#Se realiza la predicción con los datos finales de test
predicted=bestNN.predict(xFinal)
```

```python
#Se crea un archivo donde se concatenarán estos datos línea a línea
with open('./data/ytest.txt', 'w') as f:
    for p in predicted:
        print(p, file=f)
```

En resumen, la mejor arquitectura es la que utiliza los datos crudos, tiene relu de función de activación en capas ocultas, usa una tasa de aprendizaje de 0.1 y consiste en 7 capas ocultas de 10 neuronas cada una. Como se mencionó antes, la tasa de clasificación exitosa mínima es de 67.5% con una confianza de 99%, esto equivale a una probabilidad de error máxima de 32.5%. Esta probabilidad de error es un valor relativamente alto teniendo en cuenta que es un clasificador binario. Sin embargo fue el mejor resultado obtenido
